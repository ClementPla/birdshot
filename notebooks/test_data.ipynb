```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
path_text = "/home/clement/Documents/data/Birdshot/P24-170_F30.TXT"

import pandas as pd

dataset = pd.read_csv(
    path_text, header=0, encoding="unicode_escape", skiprows=51, sep="\t"
)
```

```python
chan1 = dataset["Result (nV)"]
chan2 = dataset["Result (nV).1"]
time = dataset["Time (ms)"]

plt.plot(time, chan1, label="chan1")
plt.plot(time, chan2, label="chan2")
plt.xlabel("Time (ms)")
plt.ylabel("Voltage (nV)")
plt.legend()

plt.gcf().set_size_inches(16, 8)

plt.savefig("test.png")
```

```python
path_text = "/home/clement/Documents/data/Birdshot/P24-170_Photo.TXT"

dataset = pd.read_csv(
    path_text, header=0, encoding="unicode_escape", sep="\t", skiprows=297
)
i = 0
for col in dataset.columns:
    if "Result" in col:
        i += 1
        plt.plot(dataset["Time (ms)"], dataset[col], label=f"Channel {i}")
plt.xlabel("Time (ms)")
# plt.legend()
plt.ylabel("Voltage (nV)")
fig = plt.gcf()
fig.set_size_inches(16, 8)
plt.savefig("PHOTO_TXT.png")
```
